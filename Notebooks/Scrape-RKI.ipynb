```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from lxml import html
from datetime import date
import numpy as np
```

```python
url = 'https://www.rki.de/DE/Content/InfAZ/N/Neuartiges_Coronavirus/Risikogebiete_neu.html'
```

```python
html = requests.get(url).content
soup = BeautifulSoup(html, "html.parser")
print(soup.prettify())
```

```python
RKI_geb = soup.find("div", attrs={"class": "text"}).find_all("ul")
```

```python
Risikogebiete = [i.get_text(strip=True) for i in RKI_geb]
    
```

```python
test = Risikogebiete[0].split(")")
```

```python
Euro_trans = {"Italy" : "Italien", 
             "Spain" : "Spanien", 
             "Germany" : "Deutschland", 
             "France" : "Frankreich", 
             "Switzerland" : "Schweiz",
             "UK" : "England",
             "Netherlands" : "Niederlande",
             "Austria" : "Österreich",
             "Belgium" : "Belgien", 
             "Norway" : "Norvegen", 
             "Sweden" : "Schweden", 
             "Denmark" : "Dänemark", 
             "Portugal" : "Portugal", 
             "Czechia" : "Tschechien",
             "Ireland" : "Irland",
             "Luxembourg" : "Luxemburg" ,
             "Poland" : "Polen" ,
             "Greece" : "Griechenland",
             "Finland" : "Finnland",
             "Iceland" : "Island",
             "Slovenia" : "Slowenien",
             "Romania" : "Rumänien",
             "Russia" :  "Russland",
             "Estonia" : "Estland",
             "Croatia" : "Kroatien",
             "Slovakia" : "Slowakei",
             "Serbia" : "Serbien", 
             "Bulgaria" : "Bulgarien",
             "San-Marino" : "Republik San Marino",
             "Latvia" : "Lettland",
             "Hungary" : "Ungarn",
             "Lithuania" : "Litauen",
             "Bosnia-and-Herzegovina" : "Bosnien und Herzegowina",
             "Faeroe-Islands" : "Färöer Inseln",
             "Andorra" : "Andorra",
             "North-Macedonia" : "Nordmazedonien",
             "Moldova" : "Republik Moldau",
             "Albania" : "Albanien",
             "Belarus" : "Weißrussland",
             "Malta" : "Malta",
             "Ukraine" : "Ukraine",
             "Liechtenstein" : "Liechtenstein",
             "Channel-Islands" : "Kanalinseln",
             "Monaco" : "Monaco",
             "Montenegro" : "Montenegro",
             "Gibraltar" : "Gibraltar",
             "Isle-of-Man" : "Isle of Man",
             "Vatican-City" : "Vatikanstadt"
            }
```

```python
test1 = [i for i in Euro_trans.values() for st in test if i in st]
test1
```

```python
if date.today().weekday() == 0:
    Risikogebiet_df = pd.DataFrame(Euro_trans.items(), columns = ["country", "land"]).sort_values(by=['country']).reset_index(drop = True)
    Risikogebiet_df[f"{date.today()}"] = np.where(Risikogebiet_df["land"].isin(test1), 1, 0)
    path = f"../Data/Risikogebiete_DE_week_{date.today()}.csv"
    Risikogebiet_df.to_csv(path)
else:
    week = date.today() - pd.to_timedelta((date.today().weekday()), unit='d')
    path = f"../Data/Risikogebiete_DE_week_{week}.csv"
    Risikogebiet_df = pd.read_csv(path, index_col=0)
    Risikogebiet_df[f"{date.today()}"] = np.where(Risikogebiet_df["land"].isin(test1), 1, 0)
    Risikogebiet_df.to_csv(path)
```
